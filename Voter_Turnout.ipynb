```python
import pandas as pd
import geoplot as gplt
import geopandas as gpd
import seaborn as sns
import geoplot.crs as gcrs
```

```python
cont_usa = gpd.read_file(gplt.datasets.get_path('contiguous_usa')).set_index('state')
cont_usa.head()
```

```python
obama_general_election = pd.read_csv('2008_November_General_Election.csv').set_index('state')
obama_general_election.head()
```

```python
obama_general_election = cont_usa.join(obama_general_election)
```

```python
obama_general_election.head()
```

gplt.choropleth(obama_general_election, projection=gcrs.AlbersEqualArea(), hue='VEP')

```python
obama_general_election
```

```python
ax = gplt.choropleth(obama_general_election, projection=gcrs.AlbersEqualArea(), hue='VEP Highest Office', legend=True)
ax.set_title("Voter Turnout in the 2008 General Election")
```

```python
trump_general_election = pd.read_csv('2016_November_General_Election.csv').set_index('state')
trump_general_election.head()
```

```python
trump_general_election = cont_usa.join(trump_general_election)
```
